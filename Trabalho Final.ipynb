## F Score - Piotroski

```python
import pandas as pd
import matplotlib.pyplot as plt

# contruir uma nova tabela nos quais os trimestres que NI > 0 seja 1 se não 0.
df_income = pd.read_csv('Net Income.csv', sep = ';')
df_income.head()
```

```python
def mapping(x):
    x = str(x)
    if x == '-' or x == 'nan':
        return 0
    elif float(x.replace('.','')) < 0:
        return -1
    else:
        return 1
    
def map_column(df):
    for column in df.columns:
        new_column = "mapped_" + column
        df[new_column] = df[column].apply(mapping)
    return df
```

```python
df_income = df_income.set_index('Data')
df_income = map_column(df_income)
```

```python
df_income.head()
```

```python
df_asset = pd.read_csv('Total Asset.csv', sep = ';')
df_asset = df_asset.set_index('Data')
df_asset.head()
```

```python
df_roa = pd.DataFrame()
acoes = df_asset.columns

def trans_num(x):
    x = str(x).replace('.','')
    if x == '-':
        x = 0
    else:
        x = int(x)
    return x
        
for acao in acoes:
    df_roa[acao] = df_income[acao].apply(trans_num)/df_asset[acao].apply(trans_num)
```

```python
df_roa.head()
```

```python
df_roa = map_column(df_roa)
```

```python
df_roa.head()
```

```python
df_roa.to_excel('Carteira ROA.xlsx')
```

```python
df_income.to_excel('Carteira NI.xlsx')
```

```python
df_opcash = pd.read_csv('Operating Cash Flow.csv', sep = ';')
df_opcash = df_opcash.set_index('Data')
df_opcash.head()
```

```python
df_accruals = pd.DataFrame()
```

```python
for acao in acoes:
    df_accruals[acao] = df_opcash[acao].apply(trans_num) > df_income[acao].apply(trans_num)
```

```python
df_accruals = df_accruals.astype(int)
```

```python
df_accruals.head()
```

```python
df_accruals.to_excel('Carteira Accruals.xlsx')
```
